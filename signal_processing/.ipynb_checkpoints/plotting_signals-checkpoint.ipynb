### Plotting signals
source: https://staff.fnwi.uva.nl/r.vandenboomgaard/SignalProcessing/SystemsSignals/plottingsignals.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
t = np.linspace(-0.02, 0.05, 1000)
ax.plot(t, 325 * np.sin(2*np.pi*50*t));
ax.set_xlabel('t');
ax.set_ylabel('x(t)');
ax.set_title(r'Plot of CT signal $x(t)=325 \sin(2\pi 50 t)$');
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Sequence of integers
n = np.arange(50);

# Sampling rate
dt = 0.07/50

x = np.sin(2 * np.pi * 50 * n * dt)
ax.set_xlabel('n');
ax.set_ylabel('x[n]');
ax.set_title(r'Plot of DT signal $x[n] = 325 \sin(2\pi 50 n \Delta t)$');

ax.stem(n, x, use_line_collection=True);
```

Note that in case we want to visualize a DT signal (say sound signal) with very many samples (thousands and more) we most often visualize the signal as if it were a CT signal. Only when the number of samples is small enough a stem plot will be used.

Now consider the complex valued CT signal:

$x(t) = e^{j100\pi t} = cos(100\pi t) + jsin(100\pi t)$

Plotting such a signal as if it were a real valued signal results in a Python warning that complex numbers are casted to real by discarding the imaginary parts. There are several ways to visualize a complex valued signal. The first one plots both the real and imaginary parts of the signal.

```python
plt.clf()
t = np.linspace(-0.02, 0.05, 1000)
fig, (axreal, aximag) = plt.subplots(nrows=2, sharex=True, figsize=(10, 6));
axreal.plot(t, np.exp(2j*np.pi*50*t).real );
axreal.set_xlabel('t');
axreal.set_ylabel('Re x(t)');
#plt.title(r'Real part of  \$x(t)=e^{j 100 \pi t}\$');
axreal.set_xlim([-0.02, 0.05]);

#plt.subplot(212, sharex=ax1);
aximag.plot(t, np.exp(2j*np.pi*50*t).imag );
aximag.set_xlabel('t');
aximag.set_ylabel('Im x(t)');
#plt.title(r'Imaginary part of  \$x(t)=e^{j 100 \pi t}\$');
```

Instead of plotting the real and imaginary part of a complex signal, we can also plot the magnitude and angle of the complex values.

```python
plt.clf()
t = np.linspace(-0.02, 0.05, 1000)
fig, (axreal, aximag) = plt.subplots(nrows=2, sharex=True, figsize=(10, 6));
axreal.plot(t, np.abs(np.exp(2j*np.pi*50*t)) );
axreal.set_xlabel('t');
axreal.set_ylabel(r'$|x(t)|$');
axreal.set_xlim([-0.02, 0.05]);

aximag.plot(t, np.angle(np.exp(2j*np.pi*50*t)));
aximag.set_xlabel('t');
aximag.set_ylabel(r'$\angle x(t)$');
```

Note that both plots of the complex signal are equivalent. They both do represent the same signal. Also note the seemingly strange behaviour of the phase (angle) plot. It looks quite discontinuous. But it is really not because the phase jumps from +180 degrees to -180 degrees and that is of course the same angle! This phenomenon will be seen quite a lot when plotting the phase of complex signals and functions. It is called phase wrapping. If we would allow angles outside the range from -180 to +180 we could obtain a perfectly straight line (doing just that is called phase unwrapping).

```python
complex_num = np.exp(2j*np.pi*50*1)

np.sqrt(complex_num.real**2 + complex_num.imag**2)
```
