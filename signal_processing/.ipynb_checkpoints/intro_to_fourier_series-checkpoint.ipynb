# Fourier Series

## Idea
- #### We can represent any periodic function as a weighted, infinite sum of sines and cosines
- #### Fourier Series is a way of approximating an arbitrary function (f(x)) as an infinite sum of sines and cosines of increasingly high frequency that provide an orthogonal basis for the space of solution functions.

### $f(x) = a_0 + \sum_{k=1}^{\infty} a_ksin(kx) + \sum_{k=1}^{\infty} b_mcos(mx)$

## For  a general domain $[0, L]$

### $f(x) = \frac{a_0}{2} + \sum_{k=1}^{\infty} a_ksin(\frac{2 \pi kx}{L}) + \sum_{k=1}^{\infty} b_kcos(\frac{2 \pi kx}{L})$
(Dividing by $L$ makes sine and cosine periodic on every interval of length $L$)

### $a_0 = \frac{1}{L} \int_{0}^{L}f(x)dx$
(The average value of the function over the interval $[0,L]$)

### $a_k = \frac{2}{L} \int_{0}^{L}f(x)sin(\frac{2 \pi kx}{L}) dx$
(Inner product of the function with the sine frequency, i.e., how similar is f(x) to the kth sine frequency)

### $b_k = \frac{2}{L} \int_{0}^{L}f(x)cos(\frac{2 \pi kx}{L}) dx$
(Inner product of the function with the cosine frequency, i.e., how similar is f(x) to the kth cosine frequency)

## Complex Form

### Euler's formula: $e^{i\theta} = cos(\theta) + isin(\theta)$

### $f(x) = \sum_{-\infty}^{\infty} c_ke^{ikx}$

### $c_0 = \frac{a_0}{2} $

### $c_k = \frac{b_k - iak}{2} $

### $c_{-k} = \frac{b_k + iak}{2} $

#### Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(3)
%matplotlib inline
```

#### Global Configuration 

```python
# Define the domain
x0 = 0
x1 = 100
L = (x1 - x0) / 2

# The number of fourier terms to approximate
N = 20
```

#### Util functions

```python
def create_fig():
    
    fig,ax = plt.subplots(figsize=(10,6))
    ax.grid(which='major', axis='y', color='#758D99', alpha=0.6, zorder=1)
    ax.xaxis.set_tick_params(labelsize=12)
    ax.yaxis.set_tick_params(labelsize=12)
    
    return fig, ax
```

#### Implement truncated fourier series from scratch

```python
# Let's first generate a noisy signal

# Time domain ranges from x0 to x1
domain = np.arange(x0, x1)


# Initialize the signal as an array of 0's
signal = np.zeros(shape=len(domain))


# Our signal will be made up of 100 different sine and cosine frequencies with 
# some random noise added
for k in range(0, 100):
    
    signal += np.random.randint(low=1, high=15)*np.sin(k*domain) 
    signal += np.cos(k*domain) + np.random.normal(size=len(domain))

fig, ax = create_fig()

ax.plot(domain, signal)
ax.set_xlabel('Time')
ax.set_ylabel('Signal')
```

```python
# a0 - the average value of the function over the interval [x0, x1]
a0 = (1/L)*(np.sum(signal))

# an coefficients - how similar is f(x) to the kth sine frequency
a_coefs = np.array([(1/L)*np.sum(signal*np.sin((2*np.pi*k*domain)/L)) for k in range(1, N+1)])

# bn coefficients - how similar is f(x) to the kth cosine frequency
b_coefs = np.array([(1/L)*np.sum(signal*np.cos((2*np.pi*k*domain)/L)) for k in range(1, N+1)])

b_coefs
```

```python
fs = a0/2 + np.array([a_coefs[k-1]*np.cos((2*np.pi*domain*k)/L) for k in range(1, N+1)]).sum(axis=0) + \
np.array([b_coefs[k-1]*np.sin((2*np.pi*domain*k)/L) for k in range(1, N+1)]).sum(axis=0)
```

```python
fig, ax = create_fig()

ax.plot(domain, signal)
ax.plot(domain, fs)
ax.set_xlabel('Time')
ax.set_ylabel('Signal')
```
