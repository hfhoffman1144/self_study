```python
import numpy as np
import pandas as pd
import scipy.fft as fft
from scipy import signal
import matplotlib.pyplot as plt
from typing import List
%matplotlib inline
```

#### Util functions

```python
def generate_sine_wave(freq, sample_rate:int, duration:int):
    
    # Generates samples from 0 to duration (time units) with sample_rate samples
    # per time unit (ex Hz)
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq
    
    # 2pi because np.sin takes radians
    y = np.sin((2 * np.pi) * frequencies)
    
    return x, y

def generate_signal(frequencies:List[str], sample_rate:int, duration:int, stdev:float=None):
    
    # Generate array of 0's of size sample_rate*duration
    y = np.zeros(sample_rate*duration)
    
    for freq in frequencies:
        
        t, y_tmp = generate_sine_wave(freq, sample_rate, duration)
        y += y_tmp
    
    # Add noise to the signal
    if stdev:
        
        y += np.random.normal(loc=0, scale=stdev, size=len(y))
        
    return t, y

def filter_display(t, y, fc, sample_rate, btype='low', order=5):
    
    '''Display filter design information and display filtered signal properties'''
    
    # Normalize the frequency
    w = fc / (sample_rate / 2) 

    # Get difference equation coefficients from butter worth
    b, a = signal.butter(order, w, btype)
    y_filtered = signal.filtfilt(b, a, y)
    
    # Plot original and filtered signal
    fig, ax = plt.subplots(2, 1, figsize=(20, 10), sharex=True)
    ax[0].plot(t, y, label='Original Signal')
    ax[0].legend()
    ax[1].plot(t, y_filtered, label='Filtered Signal', color='orange')
    ax[1].legend()
    plt.show()
    
    # Plot frequency and phase response
    w, h = signal.freqz(b, a, fs=sample_rate)
    h_db = 20 * np.log10(abs(h))
    h_phase = np.arctan(np.imag(h) / np.real(h))
    
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(w / max(w), h_db)
    ax.set_xlabel('Normalized Frequency')
    ax.set_ylabel('Gain (db)')
    plt.show()
    
    # Plot the fourier spectrum of the filter
    yf = fft.rfft(y_filtered)
    xf = fft.rfftfreq(len(y_filtered), 1 / SAMPLE_RATE)
    
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(xf, abs(yf))
    plt.axvline(fc, color='orange', label='Desired Cutoff Frequency')
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('Magnitude')
    ax.set_title('Fourier Spectrum of the Filter')
    ax.legend()
    plt.show()
```

#### Generate fake signal

```python
SAMPLE_RATE = 440
DURATION = 2

frequencies = [3, 5, 10, 120, 200]
stdev = 3

t, y = generate_signal(frequencies, SAMPLE_RATE, DURATION, stdev) 

y = y + 3*t
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t, y)
```

#### Lowpass filter

```python
# Cut-off frequency of the filter
fc = 10

filter_display(t, y, fc, SAMPLE_RATE, "low", 5)
```

#### Highpass filter

```python
# Cut-off frequency of the filter
fc = 50

filter_display(t, y, fc, SAMPLE_RATE, "high", 5)
```

#### Bandpass

```python
# Cut-off frequency of the filter
fc = np.array([20, 30])

filter_display(t, y, fc, SAMPLE_RATE, "band", 5)
```
