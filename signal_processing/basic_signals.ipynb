### Basic Signals
source: https://staff.fnwi.uva.nl/r.vandenboomgaard/SignalProcessing/SystemsSignals/basicsignals.html

The basic signals we will discuss in this section are seldomly encountered in real life. Nevertheless they are essential in signal processing as they are the building blocks in decomposing a signal into the basic signals (especially the complex exponential function) and characterizing systems by looking at their response to these basic signals.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```
