```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, chi2
import seaborn as sns
%matplotlib inline
```

```python
def random_uniform_sample(n:int, seed:int=10):
    '''
    Generate n samples from the uniform(0, 1) distribution
    
    Params
    ------
    n : int
    The number of samples
    
    seed : int
    The random seed
    
    Returns
    ------
    rands : List[float],
    A list of randomly sampled points from the uniform(0, 1) distribution
    '''
    
    rands = []
    for i in range(n):
        
        rn = (16*seed + 232)%(10**9)
        
        seed = rn
        rands.append(rn / (10**9))
        
    return rands

def random_normal_sample(loc:float, scale:float, n:int=1):
    
    '''
    Generate n samples from the normal(loc, scale) distribution
    
    Params
    ------
    loc : float
    The mean of the distribution
    
    scale : float
    The standard deviation of the distribution
    
    n : int
    The number of samples
    
    Returns
    ------
    Array of samples
    '''
    
    rands = random_uniform_sample(n)
    
    return norm.ppf(rands, loc=loc, scale=scale)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sample = random_normal_sample(10, 4, 1000)
sns.distplot(sample)
```
